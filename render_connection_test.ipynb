```python
sql = """
SELECT
    TO_CHAR(TO_DATE(orderdate::text, 'YYYYMMDD'), 'YYYY-MM') AS month,
    SUM(p.productunitprice * od.quantityordered) AS total_revenue
FROM
    OrderDetail od
JOIN
    Customer c ON od.customerid = c.customerid
JOIN
    Country co ON c.countryid = co.countryid
JOIN
    Region r ON co.regionid = r.regionid
JOIN
    Product p ON od.productid = p.productid
GROUP BY
    month
ORDER BY
    month
LIMIT 100;
"""

df = pd.read_sql(sql, conn)
df
```
